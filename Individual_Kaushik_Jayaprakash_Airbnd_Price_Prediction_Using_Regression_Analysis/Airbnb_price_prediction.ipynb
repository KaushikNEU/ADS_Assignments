## üìñ Theory & Background

### üß© Understanding Regression Analysis
Regression analysis is a cornerstone of **predictive modeling in data science**, used to estimate a continuous outcome variable (like price) from a set of input features (predictors).  
In the context of **Airbnb**, regression helps quantify how listing characteristics ‚Äî such as **room type**, **number of reviews**, or **location** ‚Äî influence the nightly price of a property.

---

### ‚öôÔ∏è Key Concepts
1. **Linear Regression**  
   - Assumes a straight-line relationship between predictors and response.  
   - It‚Äôs **interpretable and efficient**, allowing us to understand how much each feature contributes to price changes.  
   - However, it can **struggle with outliers** and non-linear interactions (e.g., price increases not being proportional to the number of guests).

2. **Random Forest Regressor**  
   - An **ensemble learning** method that aggregates predictions from multiple decision trees.  
   - Each tree learns different feature relationships; the combined output reduces variance and improves generalization.  
   - Excels at modeling **non-linear patterns**, **feature interactions**, and **heterogeneous data** ‚Äî common in real-world datasets like Airbnb listings.

---

### üß† Theoretical Foundation
Both approaches operate under the umbrella of **supervised learning**, where:
- The model is trained on labeled data (listings with known prices).  
- It learns to map input features (e.g., number of bedrooms, location, reviews) to output prices.  
- The trained model is then evaluated on **unseen listings** to test how well it generalizes.

This framework mirrors real-world pricing systems ‚Äî learning from past listings to estimate new ones.

---

### üîç Why This Matters
Understanding these algorithms provides two key advantages:
- **Interpretability:** Linear Regression offers transparency ‚Äî we can explain *why* a prediction was made.  
- **Performance:** Random Forest adds robustness and flexibility, capturing complex relationships that linear models miss.

Together, they showcase the trade-off between **simplicity and accuracy**, a recurring theme in modern data science practice.

## üßπ Data Preprocessing

### üéØ Objective
To **transform raw Airbnb listing data** into a clean, structured, and machine-readable format suitable for analysis and modeling.  
This step ensures that the dataset is consistent, accurate, and free from missing or misleading values before applying regression algorithms.

---

### ‚öôÔ∏è Steps and Rationale

1. **üóÇÔ∏è Data Loading**  
   - Load the dataset (`listings.csv` or `listings.csv.gz`) using **pandas**.  
   - Validate successful import and inspect dataset dimensions and column names.

2. **üí≤ Target Cleaning (`price`)**  
   - Remove currency symbols (`$`, `,`) and convert values to **float** type.  
   - Drop rows with invalid or missing price entries.  
   - Ensures the target variable is numeric and ready for regression analysis.

3. **üß© Feature Selection**  
   - Retain only relevant columns such as `room_type`, `accommodates`, `bedrooms`, `bathrooms_text`, `review_scores_rating`, and `availability_365`.  
   - Reduces noise and focuses analysis on features that logically affect price.

4. **üî¢ Parsing Text Columns (`bathrooms_text`)**  
   - Extract numeric values (e.g., `"1.5 baths"` ‚Üí `1.5`).  
   - Standardize bathroom counts for consistent numerical interpretation.

5. **üß± Handling Missing Values**  
   - Use **imputation** (median for numeric features, most frequent for categorical ones) to fill missing values.  
   - Prevents data loss from unnecessary row deletions while maintaining statistical balance.

6. **üìâ Outlier Treatment**  
   - Clip extreme prices beyond the 5th‚Äì95th percentile range.  
   - Reduces skew and prevents high-priced luxury listings from biasing the model.

7. **üè∑Ô∏è Encoding Categorical Variables**  
   - Convert non-numeric features (`room_type`, `neighbourhood_cleansed`) into numeric format using **One-Hot Encoding**.  
   - Enables regression models to interpret categorical attributes mathematically.

8. **üîÄ Data Splitting**  
   - Split the cleaned dataset into **Training (80%)** and **Testing (20%)** subsets.  
   - The training set is used for model learning, while the test set evaluates generalization performance.

---

### ‚úÖ Outcome
After preprocessing:  
- The dataset is fully numeric and free from major inconsistencies.  
- Outliers are mitigated, missing values are handled, and categorical features are encoded.  
- The data is now ready for **exploratory analysis** and **regression modeling** in the next section.


````python
import os, re, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score


# -------------------------
# 1) Load data robustly
# -------------------------
def robust_read_listings():
    candidates = ["listings.csv", "listings.csv.gz"]
    last_err = None
    for path in candidates:
        if os.path.exists(path):
            try:
                if path.endswith(".gz"):
                    return pd.read_csv(path, compression="gzip", low_memory=False)
                return pd.read_csv(path, low_memory=False)
            except Exception as e:
                last_err = e
    if last_err:
        raise last_err
    raise FileNotFoundError(
        'Could not find "listings.csv" or "listings.csv.gz" in the current directory.'
    )

df_raw = robust_read_listings()
print("Raw shape:", df_raw.shape)


# -------------------------
# 2) Select needed columns (create if missing)
# -------------------------
needed = [
    "price",
    "neighbourhood_cleansed",
    "room_type",
    "accommodates",
    "bedrooms",
    "bathrooms_text",
    "number_of_reviews",
    "review_scores_rating",
    "availability_365",
]
missing_cols = [c for c in needed if c not in df_raw.columns]
if missing_cols:
    print("Warning: Missing columns created as NaN ‚Üí", missing_cols)

df = df_raw.copy()
for c in missing_cols:
    df[c] = np.nan

df = df[needed].copy()
print("Subset shape:", df.shape)


# -------------------------
# 3) Clean target 'price' safely
# -------------------------
def clean_price_to_float(x):
    if pd.isna(x):
        return np.nan
    s = str(x)
    # strip currency symbols and commas
    s = re.sub(r"[^\d.]", "", s)
    try:
        return float(s) if s != "" else np.nan
    except:
        return np.nan

df["price"] = df["price"].apply(clean_price_to_float)

# remove clearly invalid targets
df = df[df["price"].notna() & (df["price"] > 0)]
print("After price cleaning:", df.shape)


# -------------------------
# 4) Parse bathrooms_text ‚Üí numeric
# -------------------------
def parse_bathrooms_text(x):
    if pd.isna(x):
        return np.nan
    m = re.search(r"(\d*\.?\d+)", str(x))
    if m:
        try:
            return float(m.group(1))
        except:
            return np.nan
    return np.nan

df["bathrooms"] = df["bathrooms_text"].apply(parse_bathrooms_text)
df.drop(columns=["bathrooms_text"], inplace=True)


# -------------------------
# 5) Guarded outlier clipping (only if enough rows)
# -------------------------
n_before = len(df)
if n_before >= 100:
    lo, hi = df["price"].quantile([0.05, 0.95])
    if pd.notna(lo) and pd.notna(hi) and lo < hi:
        df["price"] = df["price"].clip(lo, hi)
        print(f"Clipped price to 5‚Äì95th pct: [{lo:.2f}, {hi:.2f}]")
else:
    print("Skipping outlier clipping (too few rows).")

print("Rows after cleaning/clipping:", len(df))
if len(df) < 2:
    raise ValueError(
        "Not enough rows after cleaning. Check your data source or relax filters."
    )


# -------------------------
# 6) Features / target
# -------------------------
target = "price"
categorical_cols = ["neighbourhood_cleansed", "room_type"]
numeric_cols = [
    "accommodates",
    "bedrooms",
    "bathrooms",
    "number_of_reviews",
    "review_scores_rating",
    "availability_365",
]

# Keep only columns we actually use
keep_cols = [target] + categorical_cols + numeric_cols
df = df[keep_cols].copy()

# -------------------------
# 7) Adaptive train/test split size
# -------------------------
n = len(df)
if n >= 200:
    test_size = 0.2
elif n >= 50:
    test_size = 0.25
elif n >= 20:
    test_size = 0.33
else:
    test_size = 0.5  # maximize test split when data is tiny

X = df.drop(columns=[target])
y = df[target].astype(float)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)

print(f"Split: n={n} ‚Üí train={len(X_train)}, test={len(X_test)}, test_size={test_size}")


# -------------------------
# 8) Preprocessor (impute + encode)
#    - Categorical: most_frequent impute + OneHot
#    - Numeric: median impute
# -------------------------
preprocessor = ColumnTransformer(
    transformers=[
        (
            "cat",
            Pipeline(
                steps=[
                    ("imputer", SimpleImputer(strategy="most_frequent")),
                    ("ohe", OneHotEncoder(handle_unknown="ignore")),
                ]
            ),
            categorical_cols,
        ),
        ("num", SimpleImputer(strategy="median"), numeric_cols),
    ],
    remainder="drop",
)


# -------------------------
# 9) Models
# -------------------------
lr_model = Pipeline(
    steps=[
        ("prep", preprocessor),
        ("model", LinearRegression()),
    ]
)

rf_model = Pipeline(
    steps=[
        ("prep", preprocessor),
        ("model", RandomForestRegressor(
            n_estimators=300, random_state=42, n_jobs=-1, max_depth=None
        )),
    ]
)


# -------------------------
# 10) Fit & Evaluate with safeguards
# -------------------------
def evaluate(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    rmse = mean_squared_error(y_test, preds, squared=False)
    r2 = r2_score(y_test, preds)
    print(f"{name:>18} ‚Üí RMSE: {rmse:.2f} | R¬≤: {r2:.3f}")
    return rmse, r2

print("\nModel Performance:")
lr_scores = evaluate("Linear Regression", lr_model, X_train, y_train, X_test, y_test)
rf_scores = evaluate("Random Forest", rf_model, X_train, y_train, X_test, y_test)


# -------------------------
# 11) Simple EDA plots (optional but useful)
# -------------------------
plt.hist(df["price"], bins=50, edgecolor="k")
plt.title("Distribution of Airbnb Prices")
plt.xlabel("Price ($)")
plt.ylabel("Count")
plt.show()

# Numeric correlations (post-cleaning)
num_df = df[[c for c in numeric_cols if c in df.columns] + [target]].copy()
print("\nNumeric correlations with price:")
print(num_df.corr(numeric_only=True)[target].sort_values(ascending=False))


# -------------------------
# 12) (Optional) Feature importances for RF
#     We need to fetch the trained RF step and build the final feature names
# -------------------------
def plot_rf_feature_importances(pipeline, cat_cols, num_cols, top_k=15):
    # Extract fitted steps
    prep = pipeline.named_steps["prep"]
    rf = pipeline.named_steps["model"]

    # Build feature names from preprocessor
    ohe = prep.named_transformers_["cat"].named_steps["ohe"]
    cat_features = list(ohe.get_feature_names_out(cat_cols))
    final_features = cat_features + num_cols

    importances = rf.feature_importances_
    imp = pd.Series(importances, index=final_features).sort_values(ascending=False)
    top = imp.head(top_k)

    top.plot(kind="barh")
    plt.title("Random Forest Feature Importances (Top)")
    plt.gca().invert_yaxis()
    plt.show()
    return imp

try:
    _ = plot_rf_feature_importances(rf_model, categorical_cols, numeric_cols, top_k=15)
except Exception as e:
    print("Could not plot feature importances:", e)
````

````python
# (I) Optional: Post-model feature importances (Random Forest)
# Requires rf_model, categorical_cols, numeric_cols defined previously
from sklearn.pipeline import Pipeline

def rf_feature_importances(pipeline, cat_cols, num_cols, top_k=15):
    try:
        prep = pipeline.named_steps['prep']
        rf = pipeline.named_steps['model']
        ohe = prep.named_transformers_['cat'].named_steps['ohe']
        cat_features = list(ohe.get_feature_names_out(cat_cols))
        final_features = cat_features + num_cols

        importances = rf.feature_importances_
        order = np.argsort(importances)[::-1]
        top_idx = order[:top_k]
        top_features = [final_features[i] for i in top_idx]
        top_values = importances[top_idx]

        # Bar chart
        plt.figure(figsize=(7, 5))
        plt.barh(range(len(top_features)), top_values[::-1])
        plt.yticks(range(len(top_features)), top_features[::-1])
        plt.title('Random Forest Feature Importances (Top)')
        plt.xlabel('Importance')
        plt.tight_layout()
        plt.show()

        # Table view
        return pd.DataFrame({'feature': top_features, 'importance': top_values})
    except Exception as e:
        print("Feature importance not available:", e)
        return None

fi_table = rf_feature_importances(rf_model, categorical_cols, numeric_cols, top_k=15)
if fi_table is not None:
    display(fi_table)
````

### ‚öôÔ∏è Code Implementation

````python
# === Code Implementation: Train & Evaluate Linear Regression and Random Forest ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# ---- 1) Define target and feature groups (based on your preprocessed df) ----
target = "price"
categorical_cols = [c for c in ["neighbourhood_cleansed", "room_type"] if c in df.columns]
numeric_cols = [c for c in ["accommodates","bedrooms","bathrooms",
                            "number_of_reviews","review_scores_rating","availability_365"]
                if c in df.columns]

# Safety check
for col in [target] + categorical_cols + numeric_cols:
    if col not in df.columns:
        raise ValueError(f"Required column missing from df: {col}")

# Keep only what we need
use_cols = [target] + categorical_cols + numeric_cols
data = df[use_cols].copy()

# ---- 2) Train/Test split (adaptive size) ----
n = len(data)
if n >= 200:
    test_size = 0.2
elif n >= 50:
    test_size = 0.25
elif n >= 20:
    test_size = 0.33
else:
    test_size = 0.5

X = data.drop(columns=[target])
y = data[target].astype(float)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)
print(f"Split: n={n} ‚Üí train={len(X_train)}, test={len(X_test)}, test_size={test_size}")

# ---- 3) Preprocessor: impute + one-hot encode ----
preprocessor = ColumnTransformer(
    transformers=[
        ("cat",
         Pipeline([
             ("imputer", SimpleImputer(strategy="most_frequent")),
             ("ohe", OneHotEncoder(handle_unknown="ignore"))
         ]),
         categorical_cols),
        ("num", SimpleImputer(strategy="median"), numeric_cols),
    ],
    remainder="drop"
)

# ---- 4) Define models ----
lr_model = Pipeline([
    ("prep", preprocessor),
    ("model", LinearRegression())
])

rf_model = Pipeline([
    ("prep", preprocessor),
    ("model", RandomForestRegressor(
        n_estimators=300, random_state=42, n_jobs=-1
    ))
])

# ---- 5) Train & evaluate helper ----
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    rmse = mean_squared_error(y_test, preds, squared=False)
    r2 = r2_score(y_test, preds)
    return {"Model": name, "RMSE": rmse, "R2": r2, "Preds": preds}

# ---- 6) Fit & Evaluate both models ----
results = []
results.append(evaluate_model("Linear Regression", lr_model, X_train, y_train, X_test, y_test))
results.append(evaluate_model("Random Forest", rf_model, X_train, y_train, X_test, y_test))

# ---- 7) Show metrics table ----
metrics_df = pd.DataFrame([{k:v for k,v in r.items() if k in ["Model","RMSE","R2"]} for r in results])
metrics_df = metrics_df.sort_values(by="RMSE", ascending=True).reset_index(drop=True)
display(metrics_df.style.format({"RMSE": "{:.2f}", "R2": "{:.3f}"}))

# ---- 8) (Optional) Plot top feature importances for Random Forest ----
# This requires extracting the trained RF and feature names from the pipeline.
def plot_rf_feature_importances(pipeline, cat_cols, num_cols, top_k=15):
    try:
        prep = pipeline.named_steps["prep"]
        rf = pipeline.named_steps["model"]
        # Get OHE feature names
        ohe = prep.named_transformers_["cat"].named_steps["ohe"]
        cat_features = list(ohe.get_feature_names_out(cat_cols))
        final_features = cat_features + num_cols

        importances = rf.feature_importances_
        order = np.argsort(importances)[::-1]
        top_idx = order[:top_k]
        top_features = [final_features[i] for i in top_idx]
        top_values = importances[top_idx]

        plt.figure(figsize=(7, 5))
        plt.barh(range(len(top_features)), top_values[::-1])
        plt.yticks(range(len(top_features)), top_features[::-1])
        plt.title("Random Forest Feature Importances (Top)")
        plt.xlabel("Importance")
        plt.tight_layout()
        plt.show()

        return pd.DataFrame({"feature": top_features, "importance": top_values})
    except Exception as e:
        print("Could not compute feature importances:", e)
        return None

# Call the importance plotter on the fitted RF model
fi_table = plot_rf_feature_importances(rf_model, categorical_cols, numeric_cols, top_k=15)
if fi_table is not None:
    display(fi_table)
````


This section implements the predictive modeling workflow designed to estimate **Airbnb listing prices** based on listing attributes.  
Two regression algorithms are applied ‚Äî a **Linear Regression model** (interpretable baseline) and a **Random Forest Regressor** (non-linear ensemble model).

---

### üß± Modeling Pipeline Overview

To ensure a consistent and reproducible workflow, a **Pipeline** and **ColumnTransformer** from scikit-learn were used.

1. **Preprocessing (ColumnTransformer):**
   - **Categorical Features:** `neighbourhood_cleansed` and `room_type` were one-hot encoded using `OneHotEncoder`.
   - **Numeric Features:** Columns like `accommodates`, `bedrooms`, and `review_scores_rating` were imputed for missing values using the **median** strategy.
   - This setup ensures that any new data (e.g., unseen listings) can be processed identically during prediction.

2. **Model Definitions:**
   - **Linear Regression:**  
     A simple, interpretable baseline assuming linear relationships between features and price.  
     Pros: easy to interpret, fast to train.  
     Cons: struggles with non-linear relationships and feature interactions.
   - **Random Forest Regressor:**  
     A non-linear ensemble of decision trees trained on random feature subsets.  
     Pros: captures complex patterns, handles mixed data types well, robust to outliers.  
     Cons: less interpretable, can overfit if not tuned.

3. **Evaluation Metrics:**
   - **RMSE (Root Mean Squared Error):**  
     Measures the average prediction error in dollars ‚Äî lower is better.
   - **R¬≤ (Coefficient of Determination):**  
     Measures how much variance in price is explained by the model ‚Äî higher is better.

---

### üßÆ Model Training and Evaluation Summary

Both models were trained on **2,956 training listings** and tested on **739 unseen listings** (20% test split).  
Here are the results:

| Model | RMSE ‚Üì | R¬≤ ‚Üë | Notes |
|:--|:--:|:--:|:--|
| **Linear Regression** | 98.04 | 0.600 | Baseline performance; captures main linear relationships. |
| **Random Forest Regressor** | **87.08** | **0.685** | Improved predictive accuracy by capturing non-linear interactions. |

**Interpretation:**
- The Random Forest Regressor outperformed Linear Regression, reducing the average prediction error by roughly **11 USD per listing** and explaining **~69% of price variance**.
- This indicates that **Airbnb pricing patterns are non-linear** ‚Äî complex interactions (e.g., between location, room type, and size) are better captured by ensemble methods.
